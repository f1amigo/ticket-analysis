```python
# Import necessary libraries
import pandas as pd
import seaborn as sb
import re
```

```python
# Read csv into dataframe
df = pd.read_csv('ticket.csv')
header = df.head()
header
```

```python
# Retrieve and rename useful columns
# Replace "Which of our Members..." with the column name of the column that indicates the member's name
data = df[["Order Date", "Quantity", "Which of our Members(NTUSB) did you buy the ticket from? (Put NIL if inapplicable)"]]
data = data.rename({"Which of our Members(NTUSB) did you buy the ticket from? (Put NIL if inapplicable)": "member"}, axis=1)
display(data)
```

```python
# Clean up data
NIL = data.loc[data['member'].str.contains('nil', flags=re.IGNORECASE, regex='true')]
dataNoNIL = data.loc[~data['member'].str.contains('nil', flags=re.IGNORECASE, regex='true')]

def generalize(ser, match_name, default=None, regex=False, case=False):
    """ Search a series for text matches.
    Based on code from https://www.metasnake.com/blog/pydata-assign.html

    ser: pandas series to search
    match_name: tuple containing text to search for and text to use for normalization
    default: If no match, use this to provide a default value, otherwise use the original text
    regex: Boolean to indicate if match_name contains a  regular expression
    case: Case sensitive search

    Returns a pandas series with the matched value

    """
    seen = None
    for match, name in match_name:
        mask = ser.str.contains(match, case=case, regex=regex)
        if seen is None:
            seen = mask
        else:
            seen |= mask
        ser = ser.where(~mask, name)
    if default:
        ser = ser.where(seen, default)
    else:
        ser = ser.where(seen, ser.values)
    return ser

lookupDF = pd.DataFrame()
lookupDF['member'] = data['member'].unique()
lookup = data['member'].str.upper().value_counts()
# lookup = data['member'].str.count(flags=re.IGNORECASE)


with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3,
                       ):
    print(lookup)
```
