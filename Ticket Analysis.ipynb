```python
# Import necessary libraries
import pandas as pd
import seaborn as sb
import re
```

```python
# Read csv into dataframe
df = pd.read_csv('tickets.csv')
# names = pd.read_csv('names.csv')
header = df.head()
display(header)
```

```python
# Retrieve and rename useful columns
# Replace "Which of our Members..." with the column name of the column that indicates the member's name
data = df[["Order Date", "Quantity", "Which of our Members(NTUSB) did you buy the ticket from? (Put NIL if inapplicable)"]]
data = data.rename({"Which of our Members(NTUSB) did you buy the ticket from? (Put NIL if inapplicable)": "member"}, axis=1)
display(data)
```

```python
# Clean up data
lookup = data['member'].str.upper().value_counts().rename_axis('member').reset_index(name='counts')

with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       ):
    print(lookup)
```
